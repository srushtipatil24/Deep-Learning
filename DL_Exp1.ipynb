```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
x_train = np.random.rand(1000, 10)
y_train = np.random.randint(0,2, size=(1000, 1))
```

**Define Model**

```python
def  create_model():
    model = Sequential([
        Dense(64, activation='relu', input_shape=(10,)),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    return model
```

**SGD Optimizer**

```python
print("\n Trainning with SGD optimizer")
model = create_model()


sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(
    optimizer=sgd_optimizer,
     loss='binary_crossentropy',
     metrics=['accuracy'])

history = model.fit(x_train,y_train,epochs=10,batch_size=32)

print("Final loss:",history.history['loss'][-1])
print("Final accuracy:",history.history['accuracy'][-1])

plt.figure()
plt.plot(history.history['loss'])
plt.title('SGD Optimizer - Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

**Momentum Optimizer(SGD + momentum)**

```python
print("\n Trainning with Momentum optimizer")
model = create_model()


momentum_optimizer = tf.keras.optimizers.SGD(
    learning_rate=0.01,
    momentum=0.9
    )
model.compile(
    optimizer=momentum_optimizer,
     loss='binary_crossentropy',
     metrics=['accuracy'])

history = model.fit(x_train,y_train,epochs=10,batch_size=32)

print("Final loss:",history.history['loss'][-1])
print("Final accuracy:",history.history['accuracy'][-1])

plt.figure()
plt.plot(history.history['loss'])
plt.title('Momentum Optimizer - Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

**Nesterov Accelerated Gradient(NAG)**

```python
print("\n Trainning with NAG optimizer")
model = create_model()


nag_optimizer = tf.keras.optimizers.SGD(
    learning_rate=0.01,
    momentum=0.9,
    nesterov=True
    )
model.compile(
    optimizer=nag_optimizer,
     loss='binary_crossentropy',
     metrics=['accuracy'])

history = model.fit(x_train,y_train,epochs=10,batch_size=32)

print("Final loss:",history.history['loss'][-1])
print("Final accuracy:",history.history['accuracy'][-1])

plt.figure()
plt.plot(history.history['loss'])
plt.title('NAG Optimizer - Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

**AdaGrade Optimizer**

```python
print("\n Trainning with Adagrade optimizer")
model = create_model()


adagrade_optimizer = tf.keras.optimizers.Adagrad(learning_rate=0.01)
model.compile(
    optimizer=adagrade_optimizer,
     loss='binary_crossentropy',
     metrics=['accuracy'])

history = model.fit(x_train,y_train,epochs=10,batch_size=32)

print("Final loss:",history.history['loss'][-1])
print("Final accuracy:",history.history['accuracy'][-1])

plt.figure()
plt.plot(history.history['loss'])
plt.title('Adagrade Optimizer - Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

**RMS Prop Optimizer**

```python
print("\n Trainning with RMS prop optimizer")
model = create_model()


rms_optimizer = tf.keras.optimizers.RMSprop(
    learning_rate=0.001,
    rho = 0.9
    )
model.compile(
    optimizer=rms_optimizer,
     loss='binary_crossentropy',
     metrics=['accuracy'])

history = model.fit(x_train,y_train,epochs=10,batch_size=32)

print("Final loss:",history.history['loss'][-1])
print("Final accuracy:",history.history['accuracy'][-1])

plt.figure()
plt.plot(history.history['loss'])
plt.title('RMS prop Optimizer - Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

**Adam Optimizer**

```python
print("\n Trainning with Adam optimizer")
model = create_model()


adam_optimizer = tf.keras.optimizers.Adam(
    learning_rate=0.001,
    beta_1 = 0.9,
    beta_2 = 0.999
    )
model.compile(
    optimizer=adam_optimizer,
     loss='binary_crossentropy',
     metrics=['accuracy'])

history = model.fit(x_train,y_train,epochs=10,batch_size=32)

print("Final loss:",history.history['loss'][-1])
print("Final accuracy:",history.history['accuracy'][-1])

plt.figure()
plt.plot(history.history['loss'])
plt.title('Adam Optimizer - Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
